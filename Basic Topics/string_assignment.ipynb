```python
    # 2. Input a string and print first and last character.

s = input()
print('first character: ',s[0])
print('last character: ',s[-1])
```

```python
  # 3. Input a string and print character at index 2 and −2.

s = input()
print('character at index 2: ',s[2])
print('character at index -2: ',s[-2])
```

```python
 # 4. Input a string and print the middle character(s).

s = input()
a = len(s)

if a % 2 == 0:
    print("Middle characters:", s[a//2 - 1], s[a//2])
else:
    print("Middle character:", s[a//2])
```

```python
 # 10. Convert string to uppercase and lowercase.

s = input()
print('uppercase: ',s.upper())
print('lowercase: ',s.lower())
```

```python
 # 11. Replace all spaces in a string with -.

s = input('enter any string: ')
s = s.replace(' ','-')
print(s)
```

```python
 # 12. Check if a string starts with "A" or ends with "Z".

s = input()
if s.startswith("A"):
    print("The string starts with 'A'")
elif s.endswith("Z"):
    print("The string ends with 'Z'")
else:
    print("The string does not start with 'A' or end with 'Z'")
```

```python
 # 13. Count occurrences of a character in string using count().

s = input()
ch = input()
print(s.count(ch))
```

```python
 # 14. Find position of substring in given string.

s = input()
ch = input()
position = s.find(ch)
print(position)
```

```python
 # 15. Split a string into words and print each word separately (no loop — use index access methods).

s = input()
a = s.split()
print(a)
```

```python
 # 16. Check if input is numeric or alphabetic.

s = input()
if s.isalpha():
  print('string is alphabetic')
elif s.isnumeric():
  print('string is numeric')
else:
  print('string is alnum')
```

```python
 # 17. Swap first and last characters of string using slicing.

s = input()
s = s[-1] + s[1:-1] + s[0]
print(s)
```

```python
 # 18. Check if two strings are equal (case-insensitive).

s = input()
t = input()
if s.islower() == t.islower():
  print('two strings are equal')
else:
  print('two strings not equal')
```

```python
 # 19. Create a substring by skipping characters: s[1:len(s):2].

s = input()
print(s[1:len(s):2])
```
