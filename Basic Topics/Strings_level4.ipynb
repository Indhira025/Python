 5. What is the difference between shallow copy and deep copy in Python?
    
  - Shallow Copy
      - copy() keyword is used
      - Creates a new object,

      - But does NOT copy the inner objects.

      - Inner elements are shared between original and copy.

  - deep copy
       - Creates a new object,

       - AND copies all inner objects too.

       - No shared references.

```python
# deep copy

a = [10,50,100] # save in the Memory
b = a # deep copy

print(a)
print(b)

print(id(a))
print(id(b))

a.append(800)

print(a)
print(b)
```

```python
# shallow copy

a = [10,50,100] # save in the Memory
b = a.copy() # shallow copy

print(a)
print(b)

print(id(a))
print(id(b))

a.append(1000)

print(a)
print(b)
```

 10. How does Python manage memory internally when variables are reassigned?

    - When you reassign a variable in Python, a new object is created in memory, and the variable’s reference is updated to point to the new object.
    - Python uses reference counting and garbage collection to manage memory.
    - How It Works (Step-by-Step)
        1. Variable is assigned a value
                  x = 10

        2. Variable is reassigned
                  x = 20

        3. Python cleans unused objects

                If no variable refers to 10, Python’s garbage collector
                removes it from memory.

          This process uses:

        - Reference counting (counts how many variables point to an object)

        - Garbage collection (cleans unreferenced objects)

```python
a = 10   #  assigned first value
a = 45    # value of a updated from 10 to 45 (reassigning)
print(a)  # 45
```
