```python
# 12. Write a program to toggle the case of a given alphabet using ASCII operators:

ch = input("Enter an alphabet: ")

if 'a' <= ch <= 'z':
    print("Toggled case:", chr(ord(ch) - 32))

elif 'A' <= ch <= 'Z':
    print("Toggled case:", chr(ord(ch) + 32))

else:
    print("Not an alphabet!")
```
