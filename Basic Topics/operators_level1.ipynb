**not keyword**

The not keyword in Python is a logical operator used to obtain the negation or opposite Boolean value of an operand.

**For example,**
if False is given as an operand to not, it returns True and vice versa.


```python
x = not(10 == 10)     # 4
print(x)
```

```python
a = (9 % 4)    # 6
print(a)
```

```python
x = 3 != 3 or 5 > 4    # 7
print(x)
```

```python
x = (7 // 2)      # 8
print(x)
```

```python
a = not(3 < 2)    # 9
print(a)
```
