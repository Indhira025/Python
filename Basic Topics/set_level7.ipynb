```python
     # Coding Questions (20)
 # 1. Create a set with five numbers.

s = {12,56,89,54,65}
print(s)
```

```python
 # 2. Add an element to a set.

s = {45,85,65}
s.add(100)
print(s)
```

```python
 # 3. Remove an element from a set if present.

s = {45,21,35,65}
s.remove(21)
print(s)
```

```python
 # 4. Convert a list into a set.

l = [12,45,86,54]
print(type(l))
s = set(l)
print(s)
print(type(s))
```

```python
 # 5. Check if an element exists in a set.

s = {12,85,96,65,84}
print(89 in s)
```

```python
 # 6. Find union of two sets.

s1 = {12,53,69,87}
s2 = {56,98,78,36}
print(s1 | s2)
```

```python
 # 7. Find intersection of two sets.

s1 = {12,53,69,78}
s2 = {56,12,78,36}
print(s1 & s2)
```

```python
 # 8. Find difference between two sets.

s1 = {10,11,12,13,14}
s2 = {13,14,15,16,17}
print(s1 - s2)
```

```python
 # 9. Find symmetric difference between two sets.

s1 = {10,11,12,13,14}
s2 = {13,14,15,16,17}
print(s1 ^ s2)
```

```python
 # 10. Remove all elements from a set.

s = {15,16,17,18,19,20}
s.clear()
print(s)
```

```python
 # 11. Create a frozen set from a list.

lst = [1, 2, 3, 4]
f = frozenset(lst)

print(f)
```

```python
 # 12. Convert a string into a set of characters.

a = 'data'
s = set(a)
print(s)
```

```python
 # 13. Check if one set is subset of another.

s = {10,11,12,13,14,15}
s1 = {12,13,14}
print(s1.issubset(s))
```

```python
 # 14. Check if one set is superset of another.

s = {5,6,7,8,9,10}
s1 = {6,7,8,9,}
print(s1.issuperset(s))
```

```python
 # 15. Check if two sets are disjoint.

s = {12,13,14,15,16}
s1 = {17,18,19}
print(s1.isdisjoint(s))
```

```python
 # 16. Find length of a set.

s = {12,13,14,15}
print(len(s))
```

```python
 # 17. Pop an element from set and display result.

s = {15,14,16,18,17}
s.pop()
print(s)
```

```python
 # 18. Copy a set into another set variable.

s = {10,11,12,13}
s1 = s.copy()   # shallow copy

print(s)
print(s1)

print(id(s))
print(id(s1))

s.add(1000)

print(s)
print(s1)
```

```python
 # 19. Create a set from a tuple.

t = (12,13,14,15)
print(t)
print(type(t))
s = set(t)
print(s)
print(type(s))
```

```python
 # 20. Create two sets and show elements unique to each (symmetric diff).

s1 = {1, 2, 3}
s2 = {3, 4, 5}
print(s1 ^ s2)
```
