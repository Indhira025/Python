```python
# 1.Python program to display the given integer in reverse manner
#           sample input : 1234
#           sample output : 4321

a = 1234
s = 0

while a > 0:
  res = a % 10
  s = (s * 10) + res
  a = a // 10

print(s)
```

```python
# 2.Python program to find the sum of the digits of an integer using while loop
#              sample input : 1234
#              sample output : 10

a = 1234
s = 0

while a > 0:
  res = a % 10
  s = s + res
  a = a // 10

print(s)
```

```python
# 3.Python program to generate the prime numbers from 1 to 100
#            sample input : 7
#            sample output : is not a prime number

n = 8
c = 0

for i in range(1,n + 1):
  if n % i == 0:
    c = c + 1


if c == 2:
  print(f'{n} : is a prime number')
else:
  print(f'{n} : not a prime number')
```

```python
# 4.Write a Python program to move spaces to the front of a given string
#   ex:
#        input = sai kamal
#        output = ' saikamal'

a = 'united states usa uk'
s = 0
t = ''
for i in a:
  if i == ' ':
    s = s + 1
  else:
    t = t + i


print(s*' '+t)
```

```python
# 5.Write a Python program to count Uppercase, Lowercase, special character and numeric values in a given string.
#                  sample input : data sCience 12444
#                  sample output :
#                  upper case : 1
#                  lower case : 10
#                  numbers : 5
#                  space : 2

t = 'data sCience 12444'
upper = 0
lower = 0
sp = 0
num = 0

for i in t:
  if i.isupper():
    upper = upper + 1
  elif i.islower():
    lower = lower + 1
  elif i.isnumeric():
    num = num + 1
  else:
    sp = sp + 1

print(f'Upper letter: {upper}')
print(f'Lower letter: {lower}')
print(f'Space : {sp}')
print(f'Numerical values : {num}')
```
