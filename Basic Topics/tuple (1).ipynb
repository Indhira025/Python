```python
     # Coding Questions (no for-loop)
 # 1. Create an empty tuple.

t = ()
print(t)
```

```python
 # 2. Create a tuple with 5 integers.

t = (12,45,78,85,65)
print(t)
```

```python
 # 3. Access 3rd element of a tuple.

t = (15,45,85,96,35)
print(t[3])
```

```python
 # 4. Access last element without negative index.

t = (15,45,85,96,47)
print(t[len(t)-1])
```

```python
 # 5. Slice first three elements.

t = (15,45,85,96,47)
print(t[:3])
```

```python
 # 6. Slice every alternate value.

t = (12,45,65,69,85,32)
print(t[::2])
```

```python
 # 7. Reverse a tuple using slicing.

t = (12,54,85,78,69)
print(t[::-1])
```

```python
 # 8. Concatenate two tuples.

t1 = (12,52,78)
t2 = (52,85,36)
t = t1 + t2
print(t)
```

```python
 # 9. Replicate tuple (1,2) twice.

t = (1,2)
t = t*2
print(t)
```

```python
 # 10. Find length of a tuple.

t = (12,52,85,78)
print(len(t))
```

```python
 # 11. Convert a list to a tuple.

l = [1,5,8,6,9]
print(type(l))
t = tuple(l)
print(t)
print(type(t))
```

```python
 # 12. Convert a string "python" to tuple of characters.

s = 'python'
print(type(s))
t = tuple(s)
print(t)
print(type(t))
```

```python
 # 13. Check if value exists in tuple.

t = (12,52,78,65,85)
print(78 in t)
```

```python
 # 14. Find maximum element in tuple.

t = (12,52,85,65,45)
print(max(t))
```

```python
 # 15. Find minimum element in tuple.

t = (45,85,12,65,95)
print(min(t))
```

```python
 # 16. Count how many times 2 appears in (1,2,2,3).

t = (1,2,2,3)
print(t.count(2))
```

```python
 # 17. Find index of an element.

t = (45,85,12,65,95)
print(t.index(65))
```

```python
 # 18. Perform tuple packing.

t = 45,25.66,'data'
print(t)
print(type(t))
```

```python
 # 19. Perform tuple unpacking into variables.

t = 45,25.66,'data'
i , f , s = t
print(i)
print(f)
print(s)
```

```python
 # 20. Convert tuple to string without loops.

t = ('a', 'b', 'c')
s = ''.join(t)
print(s)
```
