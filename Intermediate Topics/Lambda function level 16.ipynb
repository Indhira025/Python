```python
# 1. Write a lambda function to return the square of a number and call it for a sample value.

f = lambda a: a * a
print(f(2))
```

```python
# 2. Write a lambda function that returns the maximum of two numbers.

f = lambda a,b: a if a > b else b
print(f(10,25))
```

```python
# 3. Write a lambda function to check whether a number is even or odd and print the result.

f = lambda n: f"even {n}" if n % 2 == 0 else f"odd {n}"
print(f(2))
```

```python
# 4. Write a lambda function that takes a list and returns a new list with each element doubled using the map function.

l = [1,2,3,4,5]
f = list(map(lambda a: a * 2,l))
print(f)
```

```python
# 5. Write a lambda function that filters and returns only the even numbers from a given list using the filter function.

a = [2,5,3,6,8]
f = list(filter(lambda x: x % 2 == 0,a))
print(f)
```
