```python
# 19. Write a program that checks if a number is prime using a for loop.

n = int(input("Enter a number: "))

if n <= 1:
    print("Not prime")
else:
    for i in range(2, n):
        if n % i == 0:
            print("Not prime")
            break
    else:
        print("Prime")
```

```python
# 20. Write a program using nested if conditions to determine if a year is a leap year.

year = int(input("Enter a year: "))

if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print(year, "is a leap year")
        else:
            print(year, "is not a leap year")
    else:
        print(year, "is a leap year")
else:
    print(year, "is not a leap year")
```
