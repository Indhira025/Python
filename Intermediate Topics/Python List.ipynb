### Question 1: Write a Python program to count the number of elements in a list without using len.


- `sample input : `
    - `var = [10,11,12,13]`
- `sample output :`
    - `result : 4`

```python
# Please write The solution Here for above question 1:

lst = input("Enter elements: ").split()

count = 0
for ch in lst:
    count += 1

print("Number of elements:", count)
```

### Question 2: Write a Python program to find the maximum and minimum values in a given list.

- `sample input : `
    - `var = [10,11,12,13,14,15]`
- `sample output :`
    - `result : max = 15 , min = 10`

```python
# Please write The solution Here for above question 2:

lst = input("Enter elements: ").split()
freq = {}

for ch in lst:
  if ch in freq:
    freq[ch] += 1
  else:
    freq[ch] = 1

print(max(freq))
print(min(freq))
```

### Question 4: Write a Python program to find the average of all the elements in a list.


- `sample input : `
    - `var = [1,2,3,4,2,5]`
- `sample output :`
    - `result : 2.833 `

```python
# Please write The solution Here for above question 4:

lst = input("Enter elements: ").split()
sum = 0
count = 0

for ch in lst:
  sum += ch
  count += 1

average = sum / count
print(average)
```

### Question 9: Write a Python program to remove the duplicates from a list

- `sample input : `
    - `var = [10,11,12,10,12,24]`
- `sample output :`
    - `result : [10,11,12,24]`

```python
# Please write The solution Here for above question 9:

lst = input('enter the elements: ').split()

unique = []

for ch in lst:
  if ch not in unique:
    unique.append(ch)

print(unique)
```

### Question 10:  Write a Python program to sort a list of strings in Ascii Order [Ascending order]


- `sample input : `
    - `var = ['a' , 'c' , 'b']`
- `sample output :`
    - `result : ['a' , 'b' , 'c']`


```python
# Please write The solution Here for above question 10:

lst = input('enter the elements: ').split()

n = len(lst)

for i in range(n):
  for j in range(i+1,n):
    if lst[i] > lst[j]:
      lst[i], lst[j] = lst[j], lst[i]
print(lst)
```
