```python
#  1.Python program to display the given integer in reverse manner
#            sample input : 1234
#            sample output : 4321

n = int(input())
rev = 0
while n > 0:
  digit = n % 10
  rev = (rev * 10) + digit
  n = n // 10

print(rev)
```

```python
# 2. Python program to find the sum of the digits of an integer using while loop
#                sample input : 1234
#                sample output : 10

n = int(input())
sum = 0
while n > 0:
  digit = n % 10
  sum = digit + sum
  n = n // 10
print(sum)
```

```python
# 3. Python program to generate the prime numbers from 1 to 100
#                    sample input : 7
#               sample output : is not a prime number

n = int(input())
c = 0
for i in range(1,100):
  if n % i == 0:
    c +=1
if c == 2:
  print(f'{n} is a prime')
else:
  print(f'{n} is not a prime')
```

```python
# 4. Write a Python program to move spaces to the front of a given string
#            input = sai kamal
#            output = ' saikamal'

n = 'sai kamal'
print(n)
b = n.replace(n[3],'')
print(' '+b)
```

```python
# 5. Write a Python program to count Uppercase, Lowercase, special character and numeric values in a given string.
#                 sample input : data sCience 12444
#                 sample output :
#                 upper case : 1
#                 lower case : 10
#                 numbers : 5
#                 space : 2

n = 'data sCience 12444'
print(n)
a = list(n)
c = 0
t = 0
d = 0
s = 0
for ch in a:
  if ch.islower():
    c += 1
  elif ch.isupper():
    t += 1
  elif ch.isnumeric():
    d += 1
  else:
    s +=1

print('lower: ',c)
print('upper: ',t)
print('numeric: ',d)
print('space: ',s)
```

```python
# 6.    L =  [ [ 12, 34, 25], [1, 45, 0] ]
#     find only the odd values from above list
#        Output = [25, 1, 45]

L =  [ [ 12, 34, 25], [1, 45, 0] ]
a=[]
for i in range(len(L[0])):
  if L[0][i] % 2 == 1:
    a.append(L[0][i])

b=[]
for j in range(len(L[1])):
  if L[1][j] % 2 == 1:
    b.append(L[1][j])

print(a+b)
```
