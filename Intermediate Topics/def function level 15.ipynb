```python
# CODING QUESTIONS
# 1. Write a function to return the square of a number.

def fun(a):
  f = a * a
  return f

fun(4)
```

```python
# 2. Write a function to check if a number is even or odd.

def f(a):
  if a % 2 == 0:
    print(f'{a} even')
  else:
    print(f'{a} odd')

f(10)
```

```python
# 3. Write a function that takes two numbers and returns the bigger one.

def largest(a,b):
  if a > b:
    print(f'largest number is {a}')
  else:
    print(f'largest number is {b}')

largest(10,20)
```

```python
# 4. Write a function to compute the factorial using recursion.

def fact(a):
  if a == 0 or a == 1:
    return 1
  else:
    return a * fact(a-1)

fact(5)
```

```python
# 5. Write a function that returns the sum of digits of a number.

def f(a):
  t = sum(int(d) for d in str(a))

  return t

f(45)
```

```python
# 6. Write a function that returns the first and last digit of a number.

def num(lst):
  return lst[0],lst[-1]

num([10,20,30])
```

```python
# 7. Write a function that checks if a number is positive negative or zero.

def fun(a):
  if a > 0:
    print(f'{a} is positive')
  elif a < 0:
    print(f'{a} is negative')
  else:
    print(f'{a} is zero')

fun(20)
```

```python
# 8. Write a function that returns the absolute difference of two numbers.

def minus(a,b):
  t = abs(a-b)        # abs is used to convert num into positive num
  return t

minus(20,50)
```

```python
# 9. Write a function that converts Celsius to Fahrenheit.

def fun(c):
  f = 9/5 * c + 32
  return f
fun(20)
```

```python
# 11. Write a function that counts vowels in a string without using loops.

def fun(s):
  t = sum(s.lower().count(a) for a in 'aeiou')
  return t

fun('string')
```

```python
# 12. Write a function that checks if a string is palindrome.

def f(s):
  t = s
  if s[::-1] == t:
    return 'palindrome'
  else:
    return 'not a palindrome'

f('141')
```

```python
# 14. Write a function that removes all spaces from a string.

def f(s):
  t = "".join(s.split())
  return t

f('hello world')
```
